```python
def moving_average(NumTimePeriod,data):
    mylist = data
    N = NumTimePeriod
    cumsum, moving_aves = [0], []
    
    for i, x in enumerate(mylist, 1):
        cumsum.append(cumsum[i-1] + x)
        if i>=N:
            moving_ave = (cumsum[i] - cumsum[i-N])/N
            moving_aves.append(moving_ave)
    return moving_aves
```

```python
len(sensor_01)
```

```python
emg_ma_01=moving_average(150,sensor_01)
emg_ma_02=moving_average(150,sensor_02)
```

```python
plt.plot(emg_ma_01,'b');
plt.title("sEMG 01 - MOVING AVERAGE");
```

```python
plt.plot(emg_ma_02,'g');
plt.title("sEMG 02 - MOVING AVERAGE");
```

```python
emg_01_max = max(emg_ma_01)
emg_02_max = max(emg_ma_02)
```

```python
plt.plot(emg_ma_01,'b');
plt.axhline(y=emg_01_max, color='r', linestyle='-')
plt.title("Sensor sEMG 01 - Procesado");
plt.xlabel('ms');
```

```python
plt.plot(emg_ma_02,'g');
plt.axhline(y=emg_02_max, color='r', linestyle='-')
plt.title("Sensor sEMG 02 - Procesado");
plt.xlabel('ms');
```
