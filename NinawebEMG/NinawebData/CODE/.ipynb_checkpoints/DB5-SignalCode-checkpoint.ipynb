Funcion que extrae los indices en los cuales termina cada estimulo

```python
def est_ind(estimulos):
    ind=[]
    for i in range(11):
        indice_inicial=estimulos[estimulos[0]==i+2].iloc[0].name #EXTRAE EL INDICE
        indice_final_anterior=indice_inicial-1
        ind.append(indice_final_anterior)
    ind.append(len(estimulos)-1)
    return ind
```

```python
# PRUEBA DE LA FUNCIÓN
indices_1=est_ind(estimulos1)
indices_1
```

Teniendo los indices en los cuales termina la accion se separa la señal en base a estos

```python
def signal_stimulus(signal,indices):
    signal_by_est=[]
    for i in range(len(indices)):
        if (i==0):
            signal_by_est.append(signal[0:indices[i]])
        else:
            signal_by_est.append(signal[indices[i-1]+1:indices[i]])
    return signal_by_est 
```

```python
# PRUEBA DE LA FUNCIÓN
signal_by_est_1_1=signal_stimulus(emg_signal_1[0],indices_1)
plt.plot(signal_by_est_1_1[7])
```

Creando una funcion que generalice el proceso de separar la señal por estimulo para todos los electrodos y los almacene en una lista de facil acceso

```python
def signaltotal_by_est(signal,estimulo):
    #Extraemos los indices de los estimulos
    indices=est_ind(estimulo)
    signaltotal=[]
    for i in range(len(signal)):
        signal_by_est=signal_stimulus(signal[i],indices)
        signaltotal.append(signal_by_est)
    return signaltotal
```

## MOVING AVERAGE - SMA

LA SEÑAL HA SIDO MUESTREADA A 200HZ

```python
def moving_average(NumTimePeriod,data):
    mylist = data
    N = NumTimePeriod
    cumsum, moving_aves = [0], []
    
    for i, x in enumerate(mylist, 1):
        cumsum.append(cumsum[i-1] + x)
        if i>=N:
            moving_ave = (cumsum[i] - cumsum[i-N])/N
            moving_aves.append(moving_ave)
    return moving_aves
```

## RMS

```python
def RMS(NumTimePeriod,data):
    mylist = data
    N = NumTimePeriod
    cumsum, moving_RMS = [0], []
    
    for i, x in enumerate(mylist, 1):
        cumsum.append(cumsum[i-1] + x*x)
        if i>=N:
            rms = np.sqrt((cumsum[i] - cumsum[i-N])/N)
            moving_RMS.append(rms)
    return moving_RMS
```

## BUTTERWORTH PASA BAJA

```python
from scipy.signal import freqz, butter, lfilter
```

```python
def butter_filter(signal,fs,cutf,grade):
    #cutf Frecuencia de corte del filtro, mientras más bajo es, más se suaviza la señal
    nyq = 0.5*fs
    low = cutf/nyq
    b,a = butter(grade,low,btype='low')
    y_filtrada=lfilter(b,a,signal)
    return y_filtrada
```

```python
emg_butter=butter_filter(emg_abs,200,5,2)
plt.figure(figsize=(20,10))
plt.plot(emg_butter,'g')
plt.title("EMG BUTTERWORTH")
```

### SEPARAR LA SEÑAL POR REPETICION

```python
def rep_by_est(estimulo,repetition):
    rep_est=[]
    indices=est_ind(estimulo)
    for i in range(len(indices)):
        if (i==0):
            rep_est.append(repetition[0:indices[i]])
        else:
            rep_est.append(repetition[indices[i-1]+1:indices[i]])
    return rep_est
```

Creando una funcion para extraer los indices por repeticion para poder separar la señal con esta

```python
def ext_ind_rep(repetitions,num_rep):
    ind_rep_total=[]
    ind_rep_start=[]
    ind_rep_end=[]
    for i in range(num_rep):
        start=repetitions[repetitions[0]==i+1].iloc[0].name
        end=repetitions[repetitions[0]==i+1].iloc[-1].name
        ind_rep_start.append(start)
        ind_rep_end.append(end)
    ind_rep_total.append(ind_rep_start)
    ind_rep_total.append(ind_rep_end)
    return ind_rep_total
```

```python
# PRUEBA DE LA FUNCIÓN
indices=ext_ind_rep(rep_est_1[0],6)
indices
```

```python
def signal_separate(signal,repetitions,num_rep):
    signal_sep=[]
    indices=ext_ind_rep(rep_est_1[0],num_rep)
    for i in range(len(indices[0])):
        signal_by_rep=signal[indices[0][i]:indices[1][i]]
        signal_sep.append(signal_by_rep)
    return signal_sep
```

### NORMALIZACIÓN DEL TIEMPO

Se requiere reescalar la señal a una cantidad igual de valores para que esta pueda promediarse, para ello una técnica es pasar los valores del eje x a porcentaje aplicando un paso de 1% o 0.5% para que no se pierda mucha información. **Sería interesante analizar el efecto de este valor en el algoritmo final**

```python
def signal_norm_time(signal,num_el):
    signal_norm=[]
    for i in range(len(signal)):
        longitud=len(signal[i])
        paso=int(longitud/num_el)
        signal_norm_i=signal[i][0:paso*num_el:paso]
        signal_norm.append(signal_norm_i)
    return signal_norm
```

```python
def med_lon(signal):
    l=0
    for i in range(len(signal)):
        l=l+len(signal[i])
    l=l/len(signal)
    return l
```

# EXTRACCIÓN DE CARACTERÍSTICAS

#### INTEGRATED EMG (IEMG)

```python
def iemg(emg):
    I=sum(np.abs(emg))
    return I
```

#### MEAN ABSOLUTE VALUE (MAV)

```python
def mav(emg):
    N=len(emg)
    mav=(1/N)*sum(np.abs(emg))
    return mav
```

#### MODIFIED MEAN ABSOLUTE VALUE 1 (MMAV1)

```python
def mmav1(emg):
    N=len(emg)
    suma=0.5*sum(np.abs(emg))+0.5*sum(np.abs(emg[int(0.25*N):int(0.75*N)]))
    mmav1=(1/N)*suma
    return mmav1
```

#### SIMPLE SQUARE INTEGRAL (SSI)

```python
def SSI(emg):
    ssi=sum(np.abs(emg)*np.abs(emg))
    return ssi
```

#### VARIANCE OF EMG (VAR)

```python
def VAR(emg):
    N=len(emg)
    var=(1/(N-1))*sum(np.abs(emg)*np.abs(emg))
    return var
```

#### ROOT MEAN SQUARE (RMS)

```python
def RMS(emg):
    N=len(emg)
    rms=np.sqrt((1/N)*sum(np.abs(emg)*np.abs(emg)))
    return rms
```

#### WAVEFORM LENGTH (WL)

```python
def WL(emg):
    N=len(emg)
    dif=emg[1:N]-emg[0:N-1]
    wl=sum(np.abs(dif))
    return wl
```

#### WILLISON AMPLITUDE (WAMP)

```python
def WAMP(emg,umbral):
    N=len(emg)
    dif=emg[1:N]-emg[0:N-1]
    dif_abs=np.abs(dif)
    s=0
    for i in range(N-1):
        if (dif_abs[i]>= umbral):
            s=s+1
    return s
```

#### LOG DETECTOR (LOG)

```python
def LOG(emg):
    N=len(emg)
    suma=sum(np.log(np.abs(emg)))
    LOG=np.exp(suma/N)
    return LOG
```

#### SLOPE SIGN CHANGE (SSC)

```python
def SSC(emg,umbral):
    N=len(emg)
    dif1=emg[1:N-1]-emg[0:N-2]
    dif2=emg[1:N-1]-emg[2:N]
    product=np.multiply(dif1,dif2)
    s=0
    for i in range(N-2):
        if (product[i]>= umbral):
            s=s+1
    return s
```

```python
SSC(media_emg,0)
```

#### ZERO CROSSING (ZC)

```python
def ZC(emg,umbral):
    N=len(emg)
    product=np.multiply(emg[0:N-1],emg[1:N])
    dif=np.abs(emg[0:N-1]-emg[1:N])
    s=0
    for i in range(N-1):
        sgn=0
        dif_cond=0
        if(product[i]>=umbral):
            sgn=1
        if(dif[i]>=umbral):
            dif_cond=1
        s = s + (sgn*dif_cond)
    return s
```

```python
ZC(media_emg,0.2)
```
